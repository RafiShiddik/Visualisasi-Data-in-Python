```python
umur_tidak_normal=[]
for i in df['age']:
  if i > 5:
    umur_tidak_normal.append("umur normal")
  else:
    umur_tidak_normal.append("umur tidak normal")
umur_tidak_normal
```

```python
jumlah_umur_normal = umur_tidak_normal.count("umur normal")

jumlah_umur_tidak_normal = umur_tidak_normal.count("umur tidak normal")


print(f"Jumlah umur normal: {jumlah_umur_normal}")
print(f"Jumlah umur tidak normal: {jumlah_umur_tidak_normal}")
```

```python
df['biaya'] = df['fare'].apply(lambda x: 'High' if x > df['fare'].mean() else 'Low')
```

```python
print(df['biaya'].value_counts())
```

```python
df['missing_survived'] = df['survived'].apply(lambda x: 'Missing' if pd.isnull(x) else 'Available')
print(df['missing_survived'].value_counts())
```

```python
df['survived'].value_counts(normalize=True) * 100
```

```python
import seaborn as sns

sns.countplot(x='survived', data=df, palette='pastel')
plt.title('Distribusi Keselamatan Penumpang')
plt.xlabel('Survived (1=Yes, 0=No)')
plt.ylabel('Jumlah Penumpang')
plt.show()
```
